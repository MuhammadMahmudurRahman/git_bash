### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Plotting in Matplotlib:

Plotting in Matplotlib can be done in two several ways either in Object Oriented Approach or pyplot/MATLAB.

Recommended way for plotting in Matplotlib:

Out of the above mentioned two ways for plotting in Matplotlib, it is recommended to use Object Oriented Approach (OO-appraoch) generally. The OO-appraoch is more convenient, specially when it needs to do more complicated plots, in addition, it is also used to do plots which are parts of a bigger project. 

On the other hand, we can use the pyplot/MATLAB approach during the time when we need to do plots for quick interactive work.

So, we can say that Matplotlib more capable and flexible way for plotting than Matlab plotting.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: 
It is the top level container for all the elements in the plot. It represents like a whole Figure and acts as a canvas where visualization of the desired plot is created.

Axes:
It can be defined as a subplot within a Figure, in other words, it is the part of the canvas where the plot is attached. It is a kind of methodology to plot from the Axes where the plot can be customized with high flexibility.

Axis:
It is a component of an Axes which sets the sacle and limits and generates the marks on the Axis which is known as ticks and strings labelling the ticks which is known as ticklabels.

Artist:
It can be defiend as what you we see on the Figure, in other words, everything which is visible on the Figure. It includes Figure, Axes and Axis and most of the Artists are tied to an Axes.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# We are going to use Object-Oriented approach below for plotting the Figure

x = np.linspace(-4, 4, 100) # linspace refers to generating linearly spaced values

# I am printing x and y for me for better understanding
print(x) 
print()

y = x**2
print(y)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Plotting of the function y = x^2')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show() # To remove the Text(0, 0.5, 'Y-axis') over the Figure. Learned it from YouTube.
```

```python
# Now we will plot the same Figure through pyplot approach

plt.plot(x, y)
plt.title('Plotting of the function y = x^2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show() # Removing the Text(0, 0.5, 'Y-axis') over the Figure. Learned it from YouTube.
```

```python
np.random.seed(15) # .seed() is used to initialize the random number generator
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2 # Given function

fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Below are the codes for creating subplots:
fig, axs = plt.subplots(1, 2,layout='constrained')
fig.suptitle('One figure containing two subplots', size = 20) # size of the figure is 20 which means we can customize the title for the Figure.

axs[0].scatter(random_data_x, random_data_y, color= 'green') # here to remember: axs o and axs 1. axs O for Scatter plot and axs 1 for Bar plot. As per difinition, under the Figure, it may contain more than one Axes.
axs[0].set_title('Scatter plot for fruits')

axs[1].bar(names, values, color = 'purple') # Learned myself, here we can command for color
axs[1].set_title('Bar plot for fruits')
plt.show() # Removing the "Text(0, 0.5, 'Y-axis')" over the Figure. Learned it from YouTube.
```
